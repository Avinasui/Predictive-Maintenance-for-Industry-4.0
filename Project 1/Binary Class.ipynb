### **Data Preprocessing**

```python
scaler = MinMaxScaler(feature_range=(0, 10))
df['Vibrations per Minute'] = scaler.fit_transform(df[['Vibrations per Minute']])
df['Vibrations per Minute'] =df['Vibrations per Minute'].astype(int)
df
```

```python
df['Vibrations per Minute'].value_counts()
```

```python
float_columns = ['noise_db', 'wpump_outlet_pressure', 'water_outlet_temp', 
                 'wpump_power', 'oilpump_power', 'oil_tank_temp']

# Apply rounding to specific columns
df[float_columns] = df[float_columns].apply(lambda x: x.round(2))
df
```

### **Feature Declaration**

```python
x=df.iloc[:,0:7] #Sensor input attributes are used as dependent variables
x
```

### **Model Training and Evaluation**

```python
x=x.drop('wpump_outlet_pressure',axis=1)
x
```

```python
# Define weights for each component 
weights = {'bearings': 0.2, 'wpump': 0.3, 'radiator': 0.7}

df['maintenance_score'] = (df['bearings'] * weights['bearings'] +
                           df['wpump'] * weights['wpump'] +
                           df['radiator'] * weights['radiator'])


df['maintenance_required'] = df['maintenance_score'] > 0.2
df['maintenance_required'] = df['maintenance_required'].astype(int)
df
```

```python
df['maintenance_score'].value_counts()
```

```python
y=df['maintenance_required'] #binary columns indicating malfunction in each component
y.value_counts()
```
